# Fingerprint Applications

```python
import cv2
import numpy as np
import matplotlib.pylab as plt

original_image=cv2.imread('fingerprint.bmp',0)

print(original_image.shape)
print(original_image.dtype)
print(type(original_image))

_=plt.imshow(original_image,'gray')
```

# Histogram

```python
_=plt.hist(original_image.ravel())
_=plt.title('histogram')
```

# Basic Global Thresholding

```python
#Basic Global Thresholding
threshold_basic,binary=cv2.threshold(original_image,45,255,cv2.THRESH_BINARY)

print('Basic Global Threshold Deeri :',threshold_basic)

_=plt.figure(figsize=(10,8))

_=plt.subplot(121),
_=plt.imshow(original_image,'gray'),
_=plt.title('original_image')

_=plt.subplot(122),
_=plt.imshow(binary,'gray'),
_=plt.title('binary')
```

# OTSUs Thresholding

```python
#OTSU's Thresholding
threshold_otsu,otsu=cv2.threshold(original_image,0,255,cv2.THRESH_OTSU)

print('OTSUs Threshold Deeri :',threshold_otsu)

_=plt.figure(figsize=(10,8))

_=plt.subplot(121),
_=plt.imshow(original_image,'gray'),
_=plt.title('original_image')

_=plt.subplot(122),
_=plt.imshow(otsu,'gray'),
_=plt.title('otsu')
```

# Triangle Thresholding

```python
#Triangle Thresholding
threshold_triangle,triangle=cv2.threshold(original_image,0,255,cv2.THRESH_TRIANGLE)

print('Triangle Threshold Deeri :',threshold_triangle)

_=plt.figure(figsize=(10,8))

_=plt.subplot(121),
_=plt.imshow(original_image,'gray'),
_=plt.title('original_image')

_=plt.subplot(122),
_=plt.imshow(triangle,'gray'),
_=plt.title('triangle')
```

```python
from skimage.io import imread,imsave, imshow
from skimage.morphology import binary_erosion,binary_dilation
from skimage.morphology import binary_opening,binary_closing
from skimage.morphology import rectangle,disk
```

# Erosion

```python
#Erosion
erosion_line=binary_erosion(otsu,rectangle(1,3))
erosion_rectangle=binary_erosion(otsu,rectangle(2,2))
erosion_disk=binary_erosion(otsu,disk(1))

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,12),sharex=True,sharey=True)

ax[0,0].imshow(otsu,cmap='gray')
ax[0,0].set_title('otsu',size=15)
ax[0,0].axis('off')

ax[0,1].imshow(erosion_line,cmap='gray')
ax[0,1].set_title('erosion_line (1,3)',size=15)
ax[0,1].axis('off')

ax[1,0].imshow(erosion_rectangle,cmap='gray')
ax[1,0].set_title('erosion_rectangle (2,2)',size=15)
ax[1,0].axis('off')

ax[1,1].imshow(erosion_disk,cmap='gray')
ax[1,1].set_title('erosion_disk (1)',size=15)
ax[1,1].axis('off')
```

# Dilation

```python
#Dilation
dilation_line=binary_dilation(otsu,rectangle(1,3))
dilation_rectangle=binary_dilation(otsu,rectangle(2,2))
dilation_disk=binary_dilation(otsu,disk(1))

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,12),sharex=True,sharey=True)

ax[0,0].imshow(otsu,cmap='gray')
ax[0,0].set_title('otsu',size=15)
ax[0,0].axis('off')

ax[0,1].imshow(dilation_line,cmap='gray')
ax[0,1].set_title('dilation_line (1,3)',size=15)
ax[0,1].axis('off')

ax[1,0].imshow(dilation_rectangle,cmap='gray')
ax[1,0].set_title('dilation_rectangle (2,2)',size=15)
ax[1,0].axis('off')

ax[1,1].imshow(dilation_disk,cmap='gray')
ax[1,1].set_title('dilation_disk (1)',size=15)
ax[1,1].axis('off')
```

# Closing

```python
#Closing
closing_line=binary_closing(otsu,rectangle(1,3))
closing_rectangle=binary_closing(otsu,rectangle(2,2))
closing_disk=binary_closing(otsu,disk(1))

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,12),sharex=True,sharey=True)

ax[0,0].imshow(otsu,cmap='gray')
ax[0,0].set_title('otsu',size=15)
ax[0,0].axis('off')

ax[0,1].imshow(closing_line,cmap='gray')
ax[0,1].set_title('closing_line (1,3)',size=15)
ax[0,1].axis('off')

ax[1,0].imshow(closing_rectangle,cmap='gray')
ax[1,0].set_title('closing_rectangle (2,2)',size=15)
ax[1,0].axis('off')

ax[1,1].imshow(closing_disk,cmap='gray')
ax[1,1].set_title('closing_disk (1)',size=15)
ax[1,1].axis('off')
```

# UIQI

```python
from sewar.full_ref import uqi

otsu[otsu>1]=1 #otsu transformed to 0,1
otsu.dtype
```

```python
# All matrixs transformed to 0,1
erosion_disk=erosion_disk.astype(otsu.dtype)
erosion_line=erosion_line.astype(otsu.dtype)
erosion_rectangle=erosion_rectangle.astype(otsu.dtype)

dilation_disk=dilation_disk.astype(otsu.dtype)
dilation_line=dilation_line.astype(otsu.dtype)
dilation_rectangle=dilation_rectangle.astype(otsu.dtype)

opening_disk=opening_disk.astype(otsu.dtype)
opening_line=opening_line.astype(otsu.dtype)
opening_rectangle=opening_rectangle.astype(otsu.dtype)

closing_disk=closing_disk.astype(otsu.dtype)
closing_line=closing_line.astype(otsu.dtype)
closing_rectangle=closing_rectangle.astype(otsu.dtype)
```

```python
filtered=[erosion_disk,erosion_line,erosion_rectangle,
          dilation_disk,dilation_line,dilation_rectangle,
          opening_disk,opening_line,opening_rectangle,
          closing_disk,closing_line,closing_rectangle]

title=['erosion_disk','erosion_line','erosion_rectangle',
       'dilation_disk','dilation_line','dilation_rectangle',
       'opening_disk','opening_line','opening_rectangle',
       'closing_disk','closing_line','closing_rectangle']

for i in range(12):
    uiqi=uqi(otsu,filtered[i])
    print ('UIQI for',title[i],':',uiqi)
```
